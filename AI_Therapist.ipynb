```python
%%writefile AI_Therapist.py
from langchain.document_loaders import TextLoader
import textwrap
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from sklearn.feature_extraction.text import TfidfVectorizer
import sentence_transformers
import faiss
from langchain.vectorstores import FAISS
from langchain.chains.question_answering import load_qa_chain
import time
import chainlit as cl
import streamlit as st

loader = TextLoader('/content/drive/MyDrive/data.txt', encoding="utf-8")
document = loader.load()

def wrap_text_preserve_newlines(text, width=110):
    lines = text.split('\n')
    wraped_lines = [textwrap.fill(line, width=width) for line in lines]
    wraped_text = '\n'.join(wraped_lines)
    return wraped_text

text_splitter = CharacterTextSplitter(chunk_size = 1000, chunk_overlap = 0)
docs = text_splitter.split_documents(document)

embeddings = HuggingFaceEmbeddings()
db = FAISS.from_documents(docs, embeddings)

st.title("AI Therapist")

if "messages" not in st.session_state:
    st.session_state.messages = []

for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.markdown(message["content"])

prompt = st.chat_input("How can I help you today?")
if prompt:

  if prompt == ("hii" or "hello"):
    st.chat_message("user").markdown(prompt)
    st.session_state.messages.append({"role": "user", "content": prompt})
    response = "Hello, I am you AI Therapist. How can I help you today?"
    with st.chat_message("assistant"):
      st.markdown(response)
    st.session_state.messages.append({"role": "assistant", "content": response})
  else:
    st.chat_message("user").markdown(prompt)
    st.session_state.messages.append({"role": "user", "content": prompt})
    doc = db.similarity_search(prompt)
    response = wrap_text_preserve_newlines(str(doc[0].page_content))
    with st.chat_message("assistant"):
      st.markdown(response)
    st.session_state.messages.append({"role": "assistant", "content": response})
```
